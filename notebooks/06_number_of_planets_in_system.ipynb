```python
# Load the dataset
data = pd.read_csv('../data/nasa_exoplanet_intelligence.csv')
data
```

```python
# Number of planets in the sistem
# System with 1 planet - 3532 etc.
n_planets_data = data['n_planets'].value_counts()
n_planets_data
```

```python
# Distribution of Planets per System
# This bar chart shows the number of detected exoplanetary systems grouped by how many planets they contain.
n_planets_data.sort_index().plot(kind='bar')
plt.xlabel('Number of planets')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.title('Distribution of planets per system')
plt.show()
```

```python
filtered_planets = data[['n_planets', 'orbital_eccentricity']].dropna()
filtered_planets
```

```python
filtered_planets = filtered_planets[filtered_planets['n_planets'] <= 5]
filtered_planets
```

```python
filtered_planets.boxplot(column='orbital_eccentricity', by='n_planets')
plt.title('Eccentricity by Number of Planets')
plt.suptitle('')
plt.xlabel('Number of planets')
plt.ylabel('Eccentricity')
plt.show()
```

### Observations

- Median eccentricity is low (close to 0) for all systems - most orbits are nearly circular.
- There is no clear decreasing trend in eccentricity as the number of planets increases.
- Systems with 1 and 2 planets show higher variability (many outliers with high eccentricity).
- Multi-planet systems (3–5) tend to have slightly more concentrated distributions, but the difference is not strong.

### Hypothesis Insight

The visual analysis does not provide strong evidence that systems with more planets have more circular orbits.

Any potential relationship appears weak and inconsistent.

```python
# Data Grouping for Statistical Testing

# In this step, the dataset is grouped by the number of planets in each system (`n_planets`).

# For each group, we extract the corresponding orbital eccentricity values and store them as separate arrays.

groups = [
    group['orbital_eccentricity'].values
    for name, group in filtered_planets.groupby('n_planets')
]
```

### ANOVA Test (One-Way)

In this step, we apply a one-way ANOVA (Analysis of Variance) test to compare the mean orbital eccentricity across different groups defined by the number of planets in a system.

The test evaluates whether there are statistically significant differences between the group means.

- **F-statistic** measures the ratio of variance between the groups to the variance within the groups.
- **P-value** indicates the probability that the observed differences occurred by chance.

If the p-value is below the chosen significance level (а = 0.01), we reject the null hypothesis and conclude that at least one group differs significantly from the others.

Otherwise, we fail to reject the null hypothesis, meaning there is no strong statistical evidence of a difference in eccentricity across systems with different numbers of planets.

```python
f_stat, p_value = f_oneway(*groups)

print("F-stat:", f_stat)
print("P-value:", p_value)
```
