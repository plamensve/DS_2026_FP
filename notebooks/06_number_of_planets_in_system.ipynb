```python
# Load the dataset
data = pd.read_csv('../data/nasa_exoplanet_intelligence.csv')
data
```

```python
# Number of planets in the sistem
# System with 1 planet - 3532 etc.
n_planets_data = data['n_planets'].value_counts()
n_planets_data
```

```python
n_planets_data.sort_index().plot(kind='bar')
plt.xlabel('Number of planets')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.title('Distribution of planets per system')
plt.show()
```
